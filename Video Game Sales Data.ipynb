# Video Game Sale Based on Genre (Global vs. JAPAN)

```python
df_genre = videogame_df.groupby('Genre')
def genreBased(region):
    xrange = np.arange(1,len(df_genre.sum())+1)
    fig,ax= plt.subplots(ncols=2,figsize=(18,6))
    df_to_plot = df_genre.sum().sort_values(by=region,ascending =False)[::-1]
    df_to_plot[region].plot(kind='barh')
    plt.title(region)
    #labels
    ax[1].set_ylabel(None)
    ax[1].tick_params(axis='both', which='major', labelsize=13)
    ax[1].set_xlabel('Total Sales(in millions)', fontsize=15,labelpad=21)
    #spines
    ax[1].spines['top'].set_visible(False)
    ax[1].spines['right'].set_visible(False)
    ax[1].grid(False)
    
    #annotations    
    for x,y in zip(np.arange(len(df_genre.sum())+1),df_genre.sum().sort_values(by=region,ascending =False)[::-1][region]):
        label = "{:}".format(y)
        labelr = round(y,2)
        plt.annotate(labelr, # this is the text
                     (y,x), # this is the point to label
                      textcoords="offset points",# how to position the text
                     xytext=(6,0), # distance from text to points (x,y)
                    ha='left',va="center")
     
    #donut chart
    theme = plt.get_cmap('Blues')
    ax[0].set_prop_cycle("color", [theme(1. * i / len(df_to_plot))for i in range(len(df_to_plot))])    
    wedges, texts,_ = ax[0].pie(df_to_plot[region], wedgeprops=dict(width=0.45), startangle=-45,labels=df_to_plot.index,
                      autopct="%.1f%%",textprops={'fontsize': 13,})

 
    plt.tight_layout()    
```

```python
genreBased('Global_Sales') #ABOVE
genreBased('JP_Sales') #BELOW 
```

```python
df_year = videogame_df.groupby('Year').sum().sort_values(by=['Year'],ascending = False)

fig,ax = plt.subplots(figsize=(20,8))
ax.plot(df_year.index,df_year['Global_Sales'],label ='Global',linewidth=2)
ax.plot(df_year.index,df_year['NA_Sales'],label ='North America',linewidth=2)
ax.plot(df_year.index,df_year['EU_Sales'],label ='Europe',linewidth=2)
ax.plot(df_year.index,df_year['JP_Sales'],label ='Japan',linewidth=2)

ax.legend(loc="center left")

ax.set_ylabel('Total Sales(in millions)', fontsize=15,labelpad=45)
ax.set_xticks(np.arange(1985,2017,1))

ax.tick_params(axis='both', which='major', labelsize=13)
ax.grid(False)

for item in ax.get_xticklabels():
    item.set_rotation(25)
```

```python
#Displot Method

df_name = ['Global','North America', 'Europe', 'Japan']
j = 0
df_lst = [videogame_df.Global_Sales,
          videogame_df.NA_Sales,
          videogame_df.EU_Sales,
          videogame_df.JP_Sales]

for i in df_lst:
    plt.figure(figsize=(20,8), dpi= 80)
    sns.distplot(i)
    plt.title('Displot Method for '+str(df_name[j])+' Sales', fontdict={'fontsize':26})
    j += 1
```
